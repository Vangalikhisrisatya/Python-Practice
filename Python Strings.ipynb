```python
str1="Hello World"
str2="Python"
```

```python
print(str1)
```

```python
#String Indexing
print(str1[4])
print(str2[2])
```

```python
#Negative Indexing
print(str1[-3])
```

```python
#string slicing
#var[start:end]
print(str1[2:8])
print(str1[-9:-4])
```

```python
print(str1[:4])
print(str1[2:])   
print(str1[:])  # prints full string
```

```python
#Modifying Strings
s1="Hello"
print("String:",s1)
list1=list(s1) #convert to list
list1.insert(5,"!")
print(list1)
s1=''.join(list1)  #converting to string
print("Modified String:",s1)
```

```python
#Concatenating strings
s1="Hello"
s2="Python"
s3=s1+" "+s2
print(s3)
```

```python
#concatenating by multiplying
print(s1*3)
```

```python
#String Formatting
name="Python Programming"
print("Welcome to %s"%name)
```

```python
name="Python"
version=3.4
print(f"{name} programming version {version}")
```

```python
#Escape Characters
print("Hello\nWorld")  #new line
print("Hello\tWorld")  #tab space
print("Hello\\World")   #print backslash
print("Hello\'World\"")   #single quote &doublequote
```

```python
#string Methods
s1="programming python"
s2="DImENsIoN"
print("capitalize:",s1.capitalize())
print("lower:",s1.lower())
print("upper:",s2.upper())
print("swapcase:",s2.swapcase())
print("casefold:",s2.casefold())
print("title:",s1.title())
```

```python
s1="  programming       "
print(s1.lstrip())
print(s1.rstrip())
print(s1.strip())
print(s1.split())
print(' '.join(s1))
```

```python
s1="Hello 123"
s2="Python"
s3="123453"
print(s1.isalnum())
print(s2.isalpha())
print(s3.isdigit())
print(s3.isnumeric())
print(s1.isspace())
print(s1.isupper())
print(s2.islower())
```
