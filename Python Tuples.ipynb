```python
t1=(1,2,3,4)
t2=("a","b","c","d")
t3=("Physics",3,True,"3+5j")
print(t1,t2,t3)
```

```python
#Accessing tuples
t1=(1,2,3,4)
t2=("a","b","c","d","e")
print(t1[1])
print(t2[1:4])
print((t2[-4]))  #negative indexing
```

```python
#using slicing
print(t2[1:3])
print(t2[-4:-2])
print(t2[:3])
print(t1[1:])
print(t1[:])
```

```python
#updating - create a new table, don't update the existing table, as tuple is immutable
t2=("a","b","c","d")
t3=("Physics",3,True,"3+5j")
t4=t2+t3
print(t4)
```

```python
#using slicing
updated_tuple=t2[:2]+t3+t2[2:]
print(updated_tuple)
```

```python
#List Comprehension
t1=(32,42,22,46,87,56,34,86,34.66,445,23)
l1_t1=list(t1)
new_l1=[i+100 for i in l1_t1]
new_t1=tuple=(new_l1)
print(t1)
print(new_t1)
```

```python
#Unpacking tuples
t1=(10,20,30)
x,y,z=t1
print(x,y,z)
```

```python
tup1 = (10,20,30)
x, *y = tup1
print ("x: ",x, "y: ", y)
```

```python
tup1 = (10,20,30, 40, 50, 60)
x, *y, z = tup1
print ("x: ",x, "y: ", y, "z: ", z)
```

```python
#loop
tup1 = (10,20,30, 40, 50, 60)
for i in tup1:
    print(i,end=" ")
```

```python
#joining tuples
t1=(12,13,45,56,67)
t2=("earth","moon","sun","stars")
t3=t1+t2
print(t3)
```

```python
t1=(12,13,45,56,67)
t2=("earth","moon","sun","stars")
l1=list(t1)
l2=list(t2)
l1.extend(l2)
t1=tuple(l1)
print(t1)
```

```python
T1 = (10,20,30,40)
T2 = ('one', 'two', 'three', 'four')
for t in T2:
    T1+=(t,)
print (T1)
```

```python
print(T1.count(20))
print(T1.index(30))
```

```python
#namedtuple - can access items using fieldnames
from collections import namedtuple
Vertex = namedtuple('Vertex', ['x', 'y'])
v = Vertex(10, 20)
print("Vertex-1:", v.x)
print("Vertex-2:", v.y)
```
