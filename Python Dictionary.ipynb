```python
marks = {"Savita":67, "Imtiaz":88, "Laxman":91, "David":49}
print ("marks dictionary before update: \n", marks)
marks1 = {"Sharad": 51, "Mushtaq": 61, "Laxman": 89}
newmarks = {**marks, **marks1}
print ("marks dictionary after update: \n", newmarks)
```

```python
marks = {"Savita":67, "Imtiaz":88, "Laxman":91, "David":49}
marks1 = {"Sharad": 51, "Mushtaq": 61, "Laxman": 89}
union=marks | marks1
print(union)
```

```python
capitals={"Andhra Pradesh":"Amaravati",
          "Telangana":"Hyderabad",
          "Tamil Nadu":"Chennai",
          "Karnataka":"Bengaluru",
          "Kerala":"Trivendrum"}
for key in capitals.keys():
    print("Keys:",key,end=" ")
print()
for value in capitals.values():
    print("Values:",value,end=" ")
print()
for key,value in capitals.items():
    print(key,":",value)
```

```python
#copy 
#shallowcopy
dict1={"name":"Alice","age":23,"skills":["Python","Data Science"]}
shallow_copy=dict1.copy()
shallow_copy["age"]=25
shallow_copy["skills"].append("Machine Learning")
print(shallow_copy)
print(dict1)
```

```python
#deepcopy
import copy
dict1={"name":"Alice","age":23,"skills":["Python","Data Science"]}
deep_copy=copy.deepcopy(dict1)
deep_copy["age"]=25
deep_copy["skills"].append("Machine Learning")
print(deep_copy)
print(dict1)
```

```python
#copy
d1={"name":"satya","age":22,"skills":"python"}
d2=d1.copy()
d2['age']=23
print(d1)
print(d2)
```

```python
students = {
    "Alice": {"age": 21, "major": "Computer Science"},
    "Bob": {"age": 20, "major": "Engineering"}}
students['Bob']['CGPA']=3.2       #adding new key-value pair
students['Charlie']={"age":23,"major":"Data Science"}  #Adding new student
print(students)
```

```python
#Accessing elements
print(students["Alice"]["major"])
print(students["Bob"]["age"])
```

```python
#using get method
s1=students.get("Alice",{}).get("major","Not found")
s2=students.get("David",{}).get("major","Not found")
print(s1)
print(s2)
```

```python
#deletion
del students['Bob']
print(students)
```

```python
for student, details in students.items():
    print(f"Student: {student}")
    for key, value in details.items():
        print(f"  {key}: {value}")
```

```python
list_comp=[(x,y) for x,y in students.items()]
print(list_comp)
```
